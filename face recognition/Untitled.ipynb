```python
import cv2
import numpy as np
import dlib
```

```python
#captures the video from primary camera 
cap = cv2.VideoCapture(0)
```

```python
# Initialize the face detector model
face_detector = dlib.get_frontal_face_detector()

# Change 0 to the index of your desired camera, or provide a video file path

# Capture frames continuously
while True:
    ret, frame = cap.read()
    frame = cv2.flip(frame, 1)

    # Convert frame to grayscale
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Detect faces in the grayscale frame
    faces = face_detector(gray)

    print(len(faces))

    # Iterate through all the detected faces and draw rectangles around each face
    for i, face in enumerate(faces, 1):
        x, y = face.left(), face.top()
        x1, y1 = face.right(), face.bottom()

        # Draw rectangle around the face
        cv2.rectangle(frame, (x, y), (x1, y1), (0, 255, 0), 2)

        # Put a label with face number
        cv2.putText(
            frame,
            "Face num " + str(i),
            (x - 10, y - 10),
            cv2.FONT_HERSHEY_SIMPLEX,
            0.7,
            (0, 0, 255),
            2,
        )

    # Display the frame
    cv2.imshow("frame", frame)

    # Check for exit condition
    if cv2.waitKey(1) & 0xFF == ord("q"):
        break

# Release the video capture and close all windows
cap.release()
cv2.destroyAllWindows()
```
